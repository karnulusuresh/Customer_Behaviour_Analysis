# Cleaning the data:

```python
# filling the null values in the review rating coloumn with median of that category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) # renamed the coloumn name.
```

```python
df.columns
```

### creating column age group

```python
labels = ["young_adult", "adult", "middle_aged", "senior"]

# quantile based discretization function used to transform continuous numerical data into discrete categorical bins.
df["age_group"] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

### create another coloumn purchase frequency days

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually': 365,
     'Quarterly': 90,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```
